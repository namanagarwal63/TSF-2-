###  **Fit and Training the Model**

```python
# Finding the optimum number of clusters for k-means classification
x = df.iloc[:, [0, 1, 2, 3]].values
```

```python
#Arbitrarily assigning a value,k=4 
km4 = KMeans(n_clusters=4) 
y_km4 = km4.fit_predict(x) 
y_km4 
```

```python
#Vaues of Centroids
km4.cluster_centers_
```

### **Elbow Plot**

**The plot will give us the optimum number of clusters for the given data set.**

```python
elpt = []
k_range = range(1, 11)
for k in k_range:
    
    km= KMeans(n_clusters=k) 
    km.fit(x) 
    elpt.append(km.inertia_) 
```

```python
elpt
```

```python
# Plotting the results onto a line graph,`allowing us to observe 'The elbow'
plt.figure (figsize=(8,4), dpi= 100, facecolor='w', edgecolor='k')
plt.plot(range(1, 11),elpt)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Error') # Within cluster sum of squares

plt.show()
```

**Now we can clearly see why it is called 'The elbow method' from the graph above; the optimum cluster is where the elbow occurs. Therefore, we will choose the number of clusters (k) as 3.**

#### **Implementing k-means using k=3**
